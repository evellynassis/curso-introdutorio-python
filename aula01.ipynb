**Controle de fluxo**

*IF,ELIF,ELSE*

A instrução ***if*** verifica uma condição que, se for ***True***, fará o código
do bloco que vem a seguir ser avaliado.



```python
#Biblioteca de número aleatórios
import random
```

```python
n1 = random.randint(1, 10)
print(n1)
```

```python
n2 = random.randint(1, 10)
print(n2)
```

```python
if(n1 != n2):
  print("Empate")
```

Uma instrução ***if*** pode ser opcionalmente seguida de um ou mais
blocos ***elif***, e um bloco ***else*** que capturará tudo mais se todas as
condições forem ***False***:

```python
if(n1 == n2):
  print('Empate')
elif(n1 > n2):
  print('Ganhou')
else:
  print('Perdeu')
```

Usando a função range()


A função range() de Python facilita gerar uma série de números. Por
exemplo, podemos usar a função range() para exibir uma sequência de
números, assim:

```python
for i in range(10):
  print(i)
```

```python
for i in range(10, 20):
  print(i)
```

```python
for i in range(10, 20, 2):
  print(i)
```

**Tratamento de exceção**

Python usa objetos especiais chamados exceções para administrar erros que
surgirem durante a execução de um programa.

```python
a = 2
b = 0
```

```python
print(a/b)
```

```python
try:
  print(a/b)
except: 
  print("Não é permitida divisão por 0")
```

```python
def divide(x, y):
    try:
      result = x / y
    except ZeroDivisionError:
      print("Divisão por zero!")
    else:
      print("O resultado é:", result)
    finally:
      print("A execução foi finalizada pela causa:", ZeroDivisionError)
```

```python
divide(a, b)
```
